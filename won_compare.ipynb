```python
## 카카오 api 를 이용한 지오코더

import requests


def getAddress(x, y):
    result = ""
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x=' + str(x) + '&y=' + str(y)
    rest_api_key = '61f1baa4fb32de937fa339702a892ff5'
    header = {'Authorization': 'KakaoAK ' + rest_api_key}
 
    r = requests.get(url, headers=header)
 
    if r.status_code == 200:
        result = r.json()["documents"][0]["region_2depth_name"]
    else:
        result = "ERROR"
    
    return result
```

```python
seoul_bus_add = []
for i in range(len(seoul_bus)):
    add = getAddress(seoul_bus['X좌표'][i], seoul_bus['Y좌표'][i])
    seoul_bus_add.append(add)
len(seoul_bus_add)
```
