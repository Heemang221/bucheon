```python
import folium
import json
import random
import numpy as np
from functools import reduce

def generate_color():
    seed = 3 * random.random()
    if seed < 1:
        i, j = 0, 1
    elif seed < 2:
        i, j = 1, 2
    else:
        i, j = 2, 0
    
    rgb = np.array([0, 0, 0])
    theta = np.pi * (seed % 1) / 2
    max_color = max(np.cos(theta), np.sin(theta))
    rgb[i] = int(255 * np.cos(theta) / max_color)
    rgb[j] = int(255 * np.sin(theta) / max_color)
    
    return '#' + reduce(lambda a, b: a + b, map(lambda x: hex(256 + x)[-2:], rgb))
```
