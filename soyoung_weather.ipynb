```python
import pandas as pd

df1 = pd.read_csv('data/weather/2016/중동_강수_201601_201612.csv')
df2 = pd.read_csv('data/weather/2016/중동_강수형태_201601_201612.csv')
df3 = pd.read_csv('data/weather/2016/중동_기온_201601_201612.csv')
df4 = pd.read_csv('data/weather/2016/중동_풍속_201601_201612.csv')
df5 = pd.read_csv('data/weather/2016/중동_풍향_201601_201612.csv')
```

```python
weather2016 = pd.concat([df1,df2,df3,df4,df5],axis=1)
weather2016.columns = ['Month','Day',"Hour","강수","강수형태","기온","풍속","풍향"]
weather2016[~weather2016["Day"].str.contains("Start")]
weather2016.drop(weather2016.tail(1).index,inplace=True)
weather2016['Month'] = weather2016['Month'].astype(int)
weather2016['Month'] = weather2016['Month'].astype(str).replace('\.\d+', '', regex=True).astype(int)


weather2016['Hour'] = weather2016['Hour']/100
weather2016['Hour'] = weather2016['Hour'].astype(int)
weather2016['Hour'] = weather2016['Hour'].astype(str).replace('\.\d+', '', regex=True).astype(int)

weather2016
```

```python
weather2016.info()
```
