```python
import pandas as pd
```

```python
academy = pd.read_csv('output/var_cluster_academy.csv')
sports = pd.read_csv('output/var_cluster_sports.csv')
budongsan = pd.read_csv('output/var_cluster_budongsan.csv')
```

```python
display(academy.head())
display(sports.head())
display(budongsan.head())
```

```python
academy['군집'].value_counts()
```

```python
academy = academy.loc[:, ['군집', '위도', '경도', '학원_5분이내거리최소']]
sports = sports.loc[:, ['군집', '스포츠_5분이내거리최대', '스포츠_5분이내거리평균']]
budongsan = budongsan.loc[:, ['군집', '부동산_5분이내거리최대']]
```

```python
total = pd.merge(left = academy, right = sports, how = 'outer', on = "군집")
total = pd.merge(left = total, right = budongsan, how = 'outer', on = "군집")
```

```python
total.head()
```

```python
weather = pd.read_csv('output/var_cluster_weather.csv')
weather = weather.loc[2:,]
weather = weather.reset_index()
weather = weather.iloc[:, 1:]
weather
```

```python
df1 = pd.DataFrame({'군집' : [], '위도' : [], '경도': [],
                    '학원_5분이내거리최소' : [], '스포츠_5분이내거리최대' : [],
                    '스포츠_5분이내거리평균' : [], '부동산_5분이내거리최대' : []})
```

```python
for i in range(len(total)):
    for j in range(10):
        df1 = df1.append(total.iloc[i, :])
df1
```

```python
df2 = pd.DataFrame({'Month' : [], '강수형태' : [], '기온': [],'풍속' : []})
```

```python
for k in range(81):
    df2 = df2.append(weather)
df2
```

```python
pred = df1.reset_index()
display(pred.head())
pred[['Month', '강수형태', '기온', '풍속']] = df2.reset_index()[['Month', '강수형태', '기온', '풍속']]
display(pred.head())
```

```python
pred = pred.iloc[:, 1:]
pred
```
