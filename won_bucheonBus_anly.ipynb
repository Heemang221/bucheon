```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# ?pd.read_csv
```

```python
bus = pd.read_csv('output/bus_num.csv')
bus = bus.iloc[:, range(1,8)]
bus.head(5)
```

```python
print(bus.isna().sum())
bus = bus.dropna(axis=0)
```

```python
bus['평균배차'] = (bus['최소배차'] + bus['최대배차']) / 2
display(bus.head())
```

```python
bus['평균배차'].describe()
```

```python
bus['첫차시간'] = pd.to_datetime(bus['첫차시간'])
bus['막차시간'] = pd.to_datetime(bus['막차시간'])
display(bus.head())
```

```python
bus.loc[bus['막차시간'].dt.hour < 4, '막차시간']
```

```python
bus.loc[bus['막차시간'].dt.hour < 4, '막차시간'] = pd.DatetimeIndex(bus[bus['막차시간'].dt.hour < 4]['막차시간']) + timedelta(days=1)
display(bus.head())
```

```python
display(bus.head())
```

```python
bus['운영시간'] = bus['막차시간'] - bus['첫차시간']
bus.head()
```

```python
bus['운영대수'] = round(bus['운영시간'] / bus['평균배차'].apply(lambda x: timedelta(minutes = x)))
bus.head()
```

```python
bus_fin = bus.loc[:, ['정류소id', '운영대수']]
bus_fin
```

```python
bus_fin = bus_fin.groupby('정류소id').sum()
bus_fin.columns = ['val']
bus_fin.head()
```

```python
bus_info = pd.read_csv('./data/버스정류소현황.csv')
is_bucheon = bus_info['시군명'] == '부천시'
bus_info = bus_info[is_bucheon]
bus_info = bus_info.loc[:, ['정류소id', '정류소명', 'WGS84위도', 'WGS84경도']]
bus_info.columns = ['정류소id', '정류소명', '위도', '경도']
display(bus_info.head(5))
```

```python
bus_fin = bus_fin.reset_index()
bus_fin.head()
```

```python
print(len(bus_fin))
bus_fin = pd.merge(left = bus_fin, right = bus_info, how = "inner", on = "정류소id")
```

```python
bus_fin.to_csv('output/bus_operate_num.csv')
```
