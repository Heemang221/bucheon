```python
import pandas as pd
bus = pd.read_csv('./data/버스정류소현황.csv')

is_bucheon = bus['시군명'] == '부천시'
bucheon_bus = bus[is_bucheon]

```

```python
import json
import folium
geo_data = "data/HangJeongDong_bucheon2.json"

with open(geo_data,encoding="utf-8") as f:
    geo_data = json.loads(f.read())

    
from folium.plugins import HeatMap
map_obj = folium.Map(location=[37.50554861215234,126.77550612183495], zoom_start=13) #tiles은 TMS URL, attr은 배경지도 정보
lat = list(bucheon_bus['WGS84위도'])
lng = list(bucheon_bus['WGS84경도'])
heatMap = HeatMap(zip(lat, lng),
                  min_opacity=0.1,
                  radius=20,  #영향 반경 
                  blur=25,     #스무딩 
                  max_zoom=4,
                  color='Red',
                  #gradient={.5: 'Yellow', .75: 'Orange', 1: 'red'}
)


folium.Choropleth(
    geo_data=geo_data,
    fill_color='Set1'
).add_to(heatMap)  

map_obj.add_child(heatMap)
map_obj
```
